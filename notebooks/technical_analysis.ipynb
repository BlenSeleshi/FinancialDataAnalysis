```python
import pandas as pd
import matplotlib.pyplot as plt
from scripts.technical_indicators import load_data, apply_technical_indicators, clean_data
```

Defining the paths for the datasets

```python
file_paths = {
    'AAPL': r'C:\Users\Blen\OneDrive\Documents\10Academy\Week1\Data\yfinance_data\AAPL_historical_data.csv',
    'AMZN': r'C:\Users\Blen\OneDrive\Documents\10Academy\Week1\Data\yfinance_data\AMZN_historical_data.csv',
    'GOOG': r'C:\Users\Blen\OneDrive\Documents\10Academy\Week1\Data\yfinance_data\GOOG_historical_data.csv',
    'META': r'C:\Users\Blen\OneDrive\Documents\10Academy\Week1\Data\yfinance_data\META_historical_data.csv',
    'MSFT': r'C:\Users\Blen\OneDrive\Documents\10Academy\Week1\Data\yfinance_data\MSFT_historical_data.csv',
    'NVDA': r'C:\Users\Blen\OneDrive\Documents\10Academy\Week1\Data\yfinance_data\NVDA_historical_data.csv',
    'TSLA': r'C:\Users\Blen\OneDrive\Documents\10Academy\Week1\Data\yfinance_data\TSLA_historical_data.csv'
}
```

Iterating over each dataset and applying the funtions we imported from our module

```python

for stock, file_path in file_paths.items():
    # Load data
    df = load_data(file_path)
    
    # Apply technical indicators
    df = apply_technical_indicators(df)
    
    # Clean the data
    df = clean_data(df)
    
    # Plotting indicators
    plt.figure(figsize=(14, 10))
    
    # Plot the closing price with SMA and EMA
    plt.subplot(311)
    plt.plot(df.index, df['Close'], label='Close Price')
    plt.plot(df.index, df['SMA_20'], label='SMA 20')
    plt.plot(df.index, df['EMA_20'], label='EMA 20')
    plt.title(f'{stock} Price and Indicators')
    plt.legend()
    
    # Plot the RSI
    plt.subplot(312)
    plt.plot(df.index, df['RSI'], label='RSI', color='orange')
    plt.axhline(70, linestyle='--', color='red')
    plt.axhline(30, linestyle='--', color='green')
    plt.title('Relative Strength Index (RSI)')
    plt.legend()
    
    # Plot the MACD and MACD Signal
    plt.subplot(313)
    plt.plot(df.index, df['MACD'], label='MACD', color='blue')
    plt.plot(df.index, df['MACD_Signal'], label='MACD Signal', color='red')
    plt.title('MACD and MACD Signal')
    plt.legend()
    
    plt.tight_layout()
    plt.show()
    
```
